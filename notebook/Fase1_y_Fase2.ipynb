### Hago una lista con los diferentes dataframes y los unimos con la funcion concat, creamos tambien un diccionario con el que traduciremos el nombre de la columnas, para que se nos haga más facil la exploración.

```python
lista_df = [df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021]

# df_concat = pd.concat(lista_df)
df_concat = pd.concat(lista_df, ignore_index=True)

df_concat.columns

tradución_columnas = {
    'CÓDIGO ÓRGÃO SUPERIOR': 'CÓDIGO ORGANISMO SUPERIOR',
    'NOME ÓRGÃO SUPERIOR': 'NOMBRE ORGANISMO SUPERIOR',
    'CÓDIGO ÓRGÃO': 'CÓDIGO ORGANISMO',
    'NOME ÓRGÃO': 'NOMBRE ORGANISMO',
    'CÓDIGO UNIDADE GESTORA': 'CÓDIGO UNIDAD GESTORA',
    'NOME UNIDADE GESTORA': 'NOMBRE UNIDAD GESTORA',
    'CATEGORIA ECONÔMICA': 'CATEGORÍA ECONÓMICA',
    'ORIGEM RECEITA': 'ORIGEN DE INGRESOS',
    'ESPÉCIE RECEITA': 'TIPO DE INGRESO',
    'DETALHAMENTO': 'DETALLE',
    'VALOR PREVISTO ATUALIZADO': 'VALOR PREVISTO ACTUALIZADO',
    'VALOR LANÇADO': 'VALOR LANZADO',
    'VALOR REALIZADO': 'VALOR REALIZADO',
    'PERCENTUAL REALIZADO': 'PORCENTAJE REALIZADO',
    'DATA LANÇAMENTO': 'FECHA DE LANZAMIENTO',
    'ANO EXERCÍCIO': 'AÑO EJERCICIO'
}

df_concat.rename(columns=tradución_columnas, inplace=True)
df_concat
```

### Observamos filas duplicadas

```python
df_concat[df_concat.duplicated()] ## Observamos que son 66 las filas duplicadas  
```

```python
df_concat.drop_duplicates(inplace=True) # Los eliminamos ya que 66 filas en nuestro conjunto de datos es una cantidad irelevante
df_concat[df_concat.duplicated()] ## Observamos que se han eliminado de manera correcta
df_concat
```

### 2. Estandarización y limpieza
- Estandarizar nombres de columnas si es necesario.

- Asegurar que los tipos de datos (fechas, valores monetarios) sean consistentes en todos los archivos.

- Tratar los valores nulos y eliminar filas o columnas irrelevantes.

Estandarizamos los puntos por las comas en los valores numéricos

```python
df_concat["CÓDIGO ORGANISMO SUPERIOR"] = df_concat["CÓDIGO ORGANISMO SUPERIOR"].astype(str).str.replace('.', ',', regex=False)
df_concat["CÓDIGO ORGANISMO"] = df_concat["CÓDIGO ORGANISMO"].astype(str).str.replace('.', ',', regex=False)
df_concat["CÓDIGO UNIDAD GESTORA"] = df_concat["CÓDIGO UNIDAD GESTORA"].astype(str).str.replace('.', ',', regex=False)
```

```python
pd.to_numeric(df_concat["VALOR LANZADO"], errors='coerce')
df_concat["VALOR LANZADO"].max()
df_concat["VALOR LANZADO"].count()
df_concat["VALOR LANZADO"].dropna()
df_concat.drop("VALOR LANZADO", axis=1, inplace=True)
df_concat.drop("PORCENTAJE REALIZADO", axis=1, inplace=True)
df_concat
```
