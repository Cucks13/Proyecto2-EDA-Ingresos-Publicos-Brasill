**Conclusión del Proyecto de Análisis Exploratorio de Datos (EDA)**

Este proyecto se enfocó en el análisis y la visualización de los ingresos públicos de Brasil desde el año 2013 hasta 2021, utilizando datos públicos que nos permitieron comprender mejor cómo se distribuyen y gestionan los recursos fiscales en el país. A lo largo de las diferentes fases del proyecto, hemos realizado una exploración exhaustiva que incluyó tanto la limpieza y preparación de los datos, como su posterior análisis visual, siempre en función de las categorías y entidades involucradas en el proceso de recaudación.

Durante la **Fase 1 y Fase 2**, nos centramos en la preparación de los datos. Esta etapa resultó ser crucial para garantizar que toda la información utilizada en los análisis posteriores estuviera estandarizada y fuera precisa. Se abordaron problemas de datos faltantes y se homogeneizaron las diferentes fuentes para poder trabajar con un conjunto de datos coherente. Adicionalmente, se realizaron muestreos para trabajar con datos representativos, evitando así la sobrecarga computacional debida al tamaño original del dataset.

En las **Fases 3 y 4**, se llevó a cabo un análisis exploratorio de datos (“EDA”) donde se exploró la **distribución de ingresos por categoría económica**, **tendencias temporales** y se identificaron **discrepancias significativas** entre lo previsto y lo realizado. Se encontraron patrones relevantes: por ejemplo, hubo una clara concentración de ciertos ingresos en unas pocas categorías, como las **Cobros Correntes**, mientras que otras presentaban una mayor distribución entre diferentes unidades gestoras. Las visualizaciones realizadas con **matplotlib** y **seaborn** proporcionaron un contexto visual para interpretar estas tendencias de manera más intuitiva, haciendo evidentes las diferencias entre lo proyectado y lo realmente recaudado.

En general, el proceso de **EDA** nos permitió profundizar en la estructura de los ingresos públicos y descubrir patrones significativos que pueden servir como base para recomendaciones políticas y fiscales. La visualización de datos fue clave para facilitar la interpretación y comunicar de manera clara las conclusiones obtenidas.

**Conclusión Final**

El análisis de los ingresos públicos de Brasil muestra una fuerte concentración en ciertas áreas de gestión y una dependencia de categorías económicas específicas. Aunque esta estructura puede aportar eficiencia, también presenta riesgos de concentración que podrían afectar la estabilidad y flexibilidad en la recaudación de ingresos. Para mejorar la resiliencia del sistema, sería recomendable diversificar la distribución de responsabilidades y fomentar un mayor equilibrio entre diferentes fuentes de ingresos, promoviendo así una gestión más robusta y adaptable.
